```python
def find_term_type(month_val, year_val):
    anom_vals = []
    if month_val <= 2:
        prev_year_rows = month_val - 3
        sub_data = nina[nina['YR'] == year_val-1]
      
        prev_sub_data = sub_data.copy()
       
        prev_year_vals = list(prev_sub_data['SEAS'].values)[prev_year_rows:]
        prev_year_anom = list(prev_sub_data['ANOM'].values)[prev_year_rows:]

        if len(prev_year_anom) > 0:
            [anom_vals.append(x) for x in prev_year_anom]
        sub_data = nina[nina['YR'] == year_val]
        for i in range(len(list(sub_data['SEAS'].values)[:12])):
            for val in list(sub_data['Months'].values)[i].split('-'):
                if val == str(month_val):
                    anom_vals.append(sub_data['ANOM'].values[i])
                    # print(sub_data['SEAS'].values[i])
            for val in list(sub_data['Months'].values)[i].split('-'):        
                if val == str(month_val - 1):
                    if sub_data['ANOM'].values[i] not in anom_vals:
                        anom_vals.append(sub_data['ANOM'].values[i])
                        # print(sub_data['SEAS'].values[i])
            for val in list(sub_data['Months'].values)[i].split('-'):        
                if val == str(month_val + 1):
                    if sub_data['ANOM'].values[i] not in anom_vals:
                        anom_vals.append(sub_data['ANOM'].values[i])
    elif month_val >= 11:
        prev_year_rows = month_val - 11
        sub_data = nina[nina['YR'] == year_val+1]
        prev_sub_data = sub_data.copy()
        # print(prev_year_rows)
        prev_year_vals = list(prev_sub_data['SEAS'].values)[:prev_year_rows]
        prev_year_anom = list(prev_sub_data['ANOM'].values)[:prev_year_rows]
        if len(prev_year_anom) > 0:
            [anom_vals.append(x) for x in prev_year_anom]
        sub_data = nina[nina['YR'] == year_val]
        for i in range(len(list(sub_data['SEAS'].values)[:12])):
            for val in list(sub_data['Months'].values)[i].split('-'):
                if val == str(month_val):
                    anom_vals.append(sub_data['ANOM'].values[i])
                    # print(sub_data['SEAS'].values[i])
            for val in list(sub_data['Months'].values)[i].split('-'):        
                if val == str(month_val - 1):
                    if sub_data['ANOM'].values[i] not in anom_vals:
                        anom_vals.append(sub_data['ANOM'].values[i])
                        # print(sub_data['SEAS'].values[i])
            # sub_data = nina[nina['YR'] == year_val+1]
            for val in list(sub_data['Months'].values)[i].split('-'):        
                if val == str(month_val + 1):
                    if sub_data['ANOM'].values[i] not in anom_vals:
                        anom_vals.append(sub_data['ANOM'].values[i])
       
    else:
        sub_data = nina[nina['YR'] == year_val]
        for i in range(len(list(sub_data['SEAS'].values)[:12])):
            for val in list(sub_data['Months'].values)[i].split('-'):
                if val == str(month_val):
                    anom_vals.append(sub_data['ANOM'].values[i])
                    # print(sub_data['SEAS'].values[i])
            for val in list(sub_data['Months'].values)[i].split('-'):        
                if val == str(month_val - 1):
                    if sub_data['ANOM'].values[i] not in anom_vals:
                        anom_vals.append(sub_data['ANOM'].values[i])
                        # print(sub_data['SEAS'].values[i])
            for val in list(sub_data['Months'].values)[i].split('-'):        
                if val == str(month_val + 1):
                    if sub_data['ANOM'].values[i] not in anom_vals:
                        anom_vals.append(sub_data['ANOM'].values[i])
                        # print(sub_data['SEAS'].values[i])
        
    sub_data = nina[nina['YR'] == year_val]
    for i in range(len(list(sub_data['SEAS'].values)[:12])):
        for val in list(sub_data['Months'].values)[i].split('-'):
            if val == str(month_val):
                anom_vals.append(sub_data['ANOM'].values[i])

    anom_avg = sum(anom_vals) / len(anom_vals)
  
    if anom_avg > 0.5:
        return 'El Niño'
        # break
    elif anom_avg < -0.5:
        return 'La Niña'
        # break
    else:   
        return 'Neutral'
```

```python
columns_to_encode = ['Ocean_Name', 'Exposure', 'Term_Type']
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoder.fit(filtered_df[columns_to_encode])
encoded_features = encoder.transform(filtered_df[columns_to_encode])
encoded_feature_names = encoder.get_feature_names_out(columns_to_encode)
encoded_df = pd.DataFrame(encoded_features, columns=encoded_feature_names)
encoded_df.head(2)
```

```python
df_final = pd.concat([filtered_df.drop(columns=columns_to_encode).reset_index(drop=True), 
                      encoded_df.reset_index(drop=True)], axis=1)
```
